```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
connect_4 = fetch_ucirepo(id=26) 
  
# data (as pandas dataframes) 
X = connect_4.data.features 
y = connect_4.data.targets 
```

```python
# combine features and targets
connect_4_df = X
connect_4_df['outcome'] = y['class']

# create subsets
connect_4_wins_losses = connect_4_df[connect_4_df['outcome'] != 'draw']
connect_4_wins = connect_4_df[connect_4_df['outcome'] == 'win']
connect_4_losses = connect_4_df[connect_4_df['outcome'] == 'loss']
connect_4_draws = connect_4_df[connect_4_df['outcome'] == 'draw']
```

```python
grid = [['a6', 'b6', 'c6', 'd6', 'e6', 'f6', 'g6'],
        ['a5', 'b5', 'c5', 'd5', 'e5', 'f5', 'g5'],
        ['a4', 'b4', 'c4', 'd4', 'e4', 'f4', 'g4'],
        ['a3', 'b3', 'c3', 'd3', 'e3', 'f3', 'g3'],
        ['a2', 'b2', 'c2', 'd2', 'e2', 'f2', 'g2'],
        ['a1', 'b1', 'c1', 'd1', 'e1', 'f1', 'g1']]
```

```python
palettes = [
    [(240,184,88), (40,32,32), (248,240,184)], # yellow-black-cream
    [(40,60,60), (0,140,140), (230,240,240)], # teals
    [(166,66,64), (233,184,114), (21,21,21)] # red-yellow-black
]
```

```python
def color(state, i):
    if state == 'x': 
        c = palettes[i][0]
    elif state == 'o': 
        c = palettes[i][1]
    else: 
        c = palettes[i][2]
    return tuple(ci/255 for ci in c)
```

```python
### draw a connect-4 board for one sample ###
```

```python
def draw_a_board(df, index, palette=0):
    """
    return the board of a sample configuration
    """
    board = df.iloc[index].to_dict()
    
    # global vars 
    rows, cols = 6, 7
    circle_radius = 1.5
    x_spacing, y_spacing = 3.1, 3.1

    # set axis limits and equal aspect ratio
    plt.axis("off")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis([0, (cols + 1) * x_spacing, 0, (rows + 1) * y_spacing])
    
    # draw connect-4 grid
    for row in range(rows):
        for col in range(cols):
            x = (col + 1) * x_spacing
            y = (rows - row) * y_spacing
            col = color(board[grid[row][col]], palette)
            circle = plt.Circle((x, y), radius=circle_radius, color=col, alpha=1, linewidth=0)
            plt.gca().add_artist(circle)

    # display
    plt.show()
```

```python
### visualize average board occupancy: transparency by state (x,o vs blank) ###
```

```python
def p_xob(df, grid):
    x_count = {cell: 0 for row in grid for cell in row}
    o_count = {cell: 0 for row in grid for cell in row}
    b_count = {cell: 0 for row in grid for cell in row}

    for index, row in df.iterrows():
        row_dict = row.to_dict()
        for key, value in row_dict.items():
            if value == 'x': x_count[key] += 1
            elif value == 'o': o_count[key] += 1
            elif value == 'b': b_count[key] += 1

    p_x = {key: val / len(df) for key,val in x_count.items()}
    p_o = {key: val / len(df) for key,val in o_count.items()}
    p_b = {key: 1 - (val / len(df)) for key,val in b_count.items()} # non-blank
    
    return [p_x, p_o, p_b]
```

```python
# separate probabilities by outcome condition
p_x_wins, p_o_wins, p_b_wins = p_xob(connect_4_wins, grid)
p_x_losses, p_o_losses, p_b_losses = p_xob(connect_4_losses, grid)
p_x_draws, p_o_draws, p_b_draws = p_xob(connect_4_draws, grid)
```

```python
def avg_fill(p_dict, c='#750505'):
    """
    return the board of occupance probabilities
    """
    # scale probs by max
    p_max = max(p_dict.values())
    p_dict_scaled = {key: val/p_max for key,val in p_dict.items()}
    
    # global vars 
    rows, cols = 6, 7
    circle_radius = 1.5
    x_spacing, y_spacing = 3.1, 3.1

    # set axis limits and equal aspect ratio
    plt.axis("off")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis([0, (cols + 1) * x_spacing, 0, (rows + 1) * y_spacing])
    
    # draw connect-4 grid
    for row in range(rows):
        for col in range(cols):
            x = (col + 1) * x_spacing
            y = (rows - row) * y_spacing
            a = p_dict_scaled[grid[row][col]]
            circle = plt.Circle((x, y), radius=circle_radius, color=c, alpha=a, linewidth=0)
            plt.gca().add_artist(circle)

    # display
    plt.show()
```

```python
# avg occupancy of 'x' by position

choice = 'all' # all, wins, losses, draws
color = '#750505'

match choice:
    case 'all':
        p_x = {cell: 0 for row in grid for cell in row}
        for k,v in p_x.items():
            p_x[k] = sum([p_x_wins[k], p_x_losses[k], p_x_draws[k]])/3
        avg_fill(p_x, color)
    case 'wins':
        avg_fill(p_x_wins, color)
    case 'losses':
        avg_fill(p_x_losses, color)
    case 'draws':
        avg_fill(p_x_draws, color)
```

```python
# avg occupancy of 'o' by position

choice = 'losses' # all, wins, losses, draws
color = '#000000'# '#bc5090'

match choice:
    case 'all':
        p_o = {cell: 0 for row in grid for cell in row}
        for k,v in p_o.items():
            p_o[k] = sum([p_o_wins[k], p_o_losses[k], p_o_draws[k]])/3
        avg_fill(p_o, color)
    case 'wins':
        avg_fill(p_o_wins, color)
    case 'losses':
        avg_fill(p_o_losses, color)
    case 'draws':
        avg_fill(p_o_draws, color)
```

```python
# avg occupancy ('x' or 'o') by position

choice = 'all' # all, wins, losses, draws
color = '#ffa600'

match choice:
    case 'all':
        p_b = {cell: 0 for row in grid for cell in row}
        for k,v in p_b.items():
            p_b[k] = sum([p_b_wins[k], p_b_losses[k], p_b_draws[k]])/3
        avg_fill(p_b, color)
    case 'wins':
        avg_fill(p_b_wins, color)
    case 'losses':
        avg_fill(p_b_losses, color)
    case 'draws':
        avg_fill(p_b_draws, color)
```

```python
def occupancy_status(p_dict,cmap):
    """
    return the board of occupance probabilities
    """
    # scale probs by max
    p_max = max(p_dict.values())
    p_dict = {key: val/p_max for key,val in p_dict.items()}
    
    rows, cols = 6, 7
    circle_radius = 1.5
    x_spacing, y_spacing = 3.1, 3.1
    colormap = cmap
    
    # set axis limits and equal aspect ratio
    plt.axis("off")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis([0, (cols + 1) * x_spacing, 0, (rows + 1) * y_spacing])
    
    # draw connect-4 grid
    for row in range(rows):
        for col in range(cols):
            x = (col + 1) * x_spacing
            y = (rows - row) * y_spacing
            val = p_dict[grid[row][col]]
            c = colormap(val)
            circle = plt.Circle((x, y), radius=circle_radius, color=c, linewidth=0)
            plt.gca().add_artist(circle)

    # display
    plt.show()
```

```python
p_b = {cell: 0 for row in grid for cell in row}
for k,v in p_b.items():
    p_b[k] = sum([p_b_wins[k], p_b_losses[k], p_b_draws[k]])/3
    
occupancy_status(p_b, cm.OrRd)
```

```python
def occupancy_status2(p_dict,cmap):
    """
    return the board of probabilities of x (blue) vs. o (red)
    """
    rows, cols = 6, 7
    circle_radius = 1.5
    x_spacing, y_spacing = 3.1, 3.1
    colormap = cmap
    norm = mcolors.TwoSlopeNorm(vmin=0, vcenter = 0.5, vmax=1)
    
    # set axis limits and equal aspect ratio
    plt.axis("off")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis([0, (cols + 1) * x_spacing, 0, (rows + 1) * y_spacing])
    
    # draw connect-4 grid
    for row in range(rows):
        for col in range(cols):
            x = (col + 1) * x_spacing
            y = (rows - row) * y_spacing
            val = p_dict[grid[row][col]]
            c = colormap(1-norm(val))
            circle = plt.Circle((x, y), radius=circle_radius, color=c, linewidth=0)
            plt.gca().add_artist(circle)

    # display
    plt.show()
```

```python
test = {cell: 0.5 for row in grid for cell in row}
test['a6'] = 0
test['a5'] = 1
test['a4'] = 4.8
test['a3'] = 3.2
test['a2'] = 1.5
test['a1'] = -1
occupancy_status2(test, cm.RdBu)
```

```python
# calculate fold change (x/o) for win outcomes
p_x_over_o_wins = {cell: 0 for row in grid for cell in row}
for k in p_x_over_o_wins.keys():
    p_x_over_o_wins[k] = p_x_wins[k] / p_o_wins[k] 

# rescale values for colormap
minp, maxp = min(p_x_over_o_wins.values()), max(p_x_over_o_wins.values())
norm = mcolors.TwoSlopeNorm(vmin=minp, vcenter = 1, vmax=maxp)
p_dict = {key: norm(val) for key,val in p_x_over_o_wins.items()}

# display board
occupancy_status2(p_dict, cm.RdBu)
```

```python
# calculate fold change (x/o) for loss outcomes
p_x_over_o_losses = {cell: 0 for row in grid for cell in row}
for k in p_x_over_o_losses.keys():
    p_x_over_o_losses[k] = p_x_losses[k] / p_o_losses[k] 

# rescale values for colormap
minp, maxp = min(p_x_over_o_losses.values()), max(p_x_over_o_losses.values())
norm = mcolors.TwoSlopeNorm(vmin=minp, vcenter = 1, vmax=maxp)
p_dict = {key: norm(val) for key,val in p_x_over_o_losses.items()}

# display board
occupancy_status2(p_dict, cm.RdBu)
```

```python
# calculate fold change (x/o) for draw outcomes
p_x_over_o_draws = {cell: 0 for row in grid for cell in row}
for k in p_x_over_o_draws.keys():
    p_x_over_o_draws[k] = p_x_draws[k] / p_o_draws[k] 

# rescale values for colormap
minp, maxp = min(p_x_over_o_draws.values()), max(p_x_over_o_draws.values())
norm = mcolors.TwoSlopeNorm(vmin=minp, vcenter = 1, vmax=maxp)
p_dict = {key: norm(val) for key,val in p_x_over_o_draws.items()}

# display board
occupancy_status2(p_dict, cm.RdBu)
```
