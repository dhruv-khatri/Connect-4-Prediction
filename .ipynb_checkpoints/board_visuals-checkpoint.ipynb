```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ast import literal_eval
import json
import random
import scipy
import sklearn
import string
from sklearn import linear_model
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
connect_4 = fetch_ucirepo(id=26) 
  
# data (as pandas dataframes) 
X = connect_4.data.features 
y = connect_4.data.targets 
```

```python
connect4_data = X
connect4_data['outcome'] = y['class']
connect4_data = connect4_data[connect4_data['outcome'] != 'draw']
```

```python
grid = [['a6', 'b6', 'c6', 'd6', 'e6', 'f6', 'g6'],
        ['a5', 'b5', 'c5', 'd5', 'e5', 'f5', 'g5'],
        ['a4', 'b4', 'c4', 'd4', 'e4', 'f4', 'g4'],
        ['a3', 'b3', 'c3', 'd3', 'e3', 'f3', 'g3'],
        ['a2', 'b2', 'c2', 'd2', 'e2', 'f2', 'g2'],
        ['a1', 'b1', 'c1', 'd1', 'e1', 'f1', 'g1']]
```

```python
palettes = [
    [(240,184,88), (40,32,32), (248,240,184)], # yellow-black-cream
    [(40,60,60), (0,140,140), (230,240,240)], # teals
    [(166,66,64), (233,184,114), (21,21,21)] # red-yellow-black
]
```

```python
def color(state, i):
    if state == 'x': 
        c = palettes[i][0]
    elif state == 'o': 
        c = palettes[i][1]
    else: 
        c = palettes[i][2]
    return tuple(ci/255 for ci in c)
```

```python
### draw a connect-4 board for one sample ###
```

```python
def draw_a_board(df, index, palette=0):
    """
    return the board of a sample configuration
    """
    board = df.iloc[index].to_dict()
    
    # global vars 
    rows, cols = 6, 7
    circle_radius = 1.5
    x_spacing, y_spacing = 3.1, 3.1

    # set axis limits and equal aspect ratio
    plt.axis("off")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis([0, (cols + 1) * x_spacing, 0, (rows + 1) * y_spacing])
    
    # draw connect-4 grid
    for row in range(rows):
        for col in range(cols):
            x = (col + 1) * x_spacing
            y = (rows - row) * y_spacing
            col = color(board[grid[row][col]], palette)
            circle = plt.Circle((x, y), radius=circle_radius, color=col, alpha=1, linewidth=0)
            plt.gca().add_artist(circle)

    # display
    plt.show()
```

```python
draw_a_board(connect4_data, 550, 0)
```

```python
draw_a_board(connect4_data, 550, 1)
```

```python
draw_a_board(connect4_data, 550, 2)
```

```python
### visualize average board occupancy: transparency by state (x,o vs blank) ###
```

```python
x_counts = {cell: 0 for row in grid for cell in row}
o_counts = {cell: 0 for row in grid for cell in row}
b_counts = {cell: 0 for row in grid for cell in row}
```

```python
for index, row in connect4_data.iterrows():
    row_dict = row.to_dict()
    for key, value in row_dict.items():
        if value == 'x': x_counts[key] += 1
        elif value == 'o': o_counts[key] += 1
        elif value == 'b': b_counts[key] += 1
```

```python
x_true_p = {key: val / len(connect4_data) for key,val in x_counts.items()}
o_true_p = {key: val / len(connect4_data) for key,val in o_counts.items()}
b_true_p = {key: val / len(connect4_data) for key,val in b_counts.items()}
```

```python
not_b_true_p = {key: 1-val for key,val in b_true_p.items()}
```

```python
def avg_fill(df, p_dict, c='#750505'):
    """
    return the board of occupance probabilities
    """
    # scale probs by max
    p_max = max(p_dict.values())
    p_dict_scaled = {key: val/p_max for key,val in p_dict.items()}
    
    # global vars 
    rows, cols = 6, 7
    circle_radius = 1.5
    x_spacing, y_spacing = 3.1, 3.1

    # set axis limits and equal aspect ratio
    plt.axis("off")
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis([0, (cols + 1) * x_spacing, 0, (rows + 1) * y_spacing])
    
    # draw connect-4 grid
    for row in range(rows):
        for col in range(cols):
            x = (col + 1) * x_spacing
            y = (rows - row) * y_spacing
            a = p_dict_scaled[grid[row][col]]
            circle = plt.Circle((x, y), radius=circle_radius, color=c, alpha=a, linewidth=0)
            plt.gca().add_artist(circle)

    # display
    plt.show()
```

```python
# avg board occupancy by position
avg_fill(connect4_data, not_b_true_p)
```

```python
# avg occupancy of 'x' by position
avg_fill(connect4_data, x_true_p, '#ffa600')
```

```python
# avg occupancy of 'o' by position
avg_fill(connect4_data, o_true_p, '#bc5090')
```
